```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Home Page
def home_page():
    print("Welcome to the Streamlit Web App!")
    print("""
    This is a simple web app built with Python.
    You can interact with different sections like User Info, Data Tables, and Visualizations.
    The app also lets you upload files and perform some basic operations.
    """)

# User Info Page
def user_info_page():
    print("Tell Us About Yourself")
    name = input("What is your name? ")
    age = int(input("How old are you? "))

    print(f"Hello {name}, you are {age} years old! ðŸ‘‹")

# Sample Data Page
def sample_data_page():
    print("Sample Data Table")
    data = {
        'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 45],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
    }

    df = pd.DataFrame(data)
    print("Here is a sample data table:")
    print(df)

# Visualization Page
def visualization_page():
    print("Data Visualization with Matplotlib")

    # Generating random data for plot
    x = np.linspace(0, 10, 100)
    y = np.sin(x)

    fig, ax = plt.subplots()
    ax.plot(x, y, label="Sine Wave", color='orange')
    ax.set_title("Sine Wave Plot")
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.legend()

    plt.show()

# File Upload Page
def file_upload_page():
    print("Upload Your File")

    # In Google Colab, we can't upload files directly like Streamlit.
    # Instead, you can use file upload from Colab
    from google.colab import files
    uploaded = files.upload()

    if uploaded:
        import io
        # Read the uploaded CSV file into a pandas dataframe
        df_uploaded = pd.read_csv(io.BytesIO(uploaded['your_file.csv']))
        print("Here is the uploaded file:")
        print(df_uploaded)

# Main function to run the app and simulate navigation
def main():
    print("Choose a page:")
    print("1. Home")
    print("2. User Info")
    print("3. Sample Data")
    print("4. Visualization")
    print("5. File Upload")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        home_page()
    elif choice == "2":
        user_info_page()
    elif choice == "3":
        sample_data_page()
    elif choice == "4":
        visualization_page()
    elif choice == "5":
        file_upload_page()
    else:
        print("Invalid choice!")

if __name__ == '__main__':
    main()
```
